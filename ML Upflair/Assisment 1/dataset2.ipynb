# 20
 Plot a bar chart to visualize the number of employees in each department using Matplotlib.

 ---


```python
import pandas as pd
import matplotlib.pyplot as plt

# Read data from CSV file
df = pd.read_csv("department_salary.csv")

# Count employees in each department
dept_counts = df["Department"].value_counts()

# Plot the bar chart
plt.figure(figsize=(8, 5))
plt.bar(dept_counts.index, dept_counts.values, color=['pink', 'orange', 'black'])

# Add labels and title
plt.xlabel("Department")
plt.ylabel("Number of Employees")
plt.title("Number of Employees in Each Department")

# Show the plot
plt.show()
```

# 21
Filter out employees with a salary greater than 60,000 using Pandas.

---

```python
# Filter employees with salary > 60000
high_salary_employees = df[df["Salary"] > 60000]

# Print the result
print(high_salary_employees)
```

# 22
Create a line graph for the salary trend by department.

---

```python
# Calculate average salary per department
avg_salary_by_dept = df.groupby("Department")["Salary"].mean()

# Plot the line graph
plt.figure(figsize=(8, 5))
plt.plot(avg_salary_by_dept.index, avg_salary_by_dept.values, marker='o', linestyle='--', color='black')

# Add labels and title
plt.xlabel("Department")
plt.ylabel("Average Salary")
plt.title("Salary Trend by Department")

# Show the plot
plt.show()
```
