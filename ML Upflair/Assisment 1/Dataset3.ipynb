# 23
Using NumPy, calculate the mean score of each subject.

---

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("student_data.csv")

# Calculate mean of each subject
mean_scores = df.iloc[:, 1:].mean().to_dict()

# Print mean scores
print("Mean Scores:", mean_scores)
```

# 24
Create a Pandas DataFrame for this dataset and calculate the average score of each student.

---

```python
# Calculate average score per student
df["Average Score"] = df.iloc[:, 1:].mean(axis=1)

# Print the updated DataFrame
print(df[["Student Name", "Average Score"]])
```

# 25
Find the student with the highest overall score.

---

```python
# Calculate total score for each student
df["Total Score"] = df.iloc[:, 1:5].sum(axis=1)

# Find the student with the highest score
top_student = df.loc[df["Total Score"].idxmax(), "Student Name"]
print("Top Student:", top_student)
```

# 26
Plot a scatter plot using Matplotlib to visualize Math vs Science scores.

---

```python
plt.figure(figsize=(6, 4))
plt.scatter(df["Math"], df["Science"], color='red', label="Math vs Science")

# Add labels
plt.xlabel("Math Scores")
plt.ylabel("Science Scores")
plt.title("Scatter Plot: Math vs Science Scores")
plt.legend()
plt.show()
```

# 27
Standardize the scores using NumPy by subtracting the mean and dividing by the standard deviation.

---

```python
# Standardizing: (X - mean) / std
standardized_scores = (df.iloc[:, 1:5] - df.iloc[:, 1:5].mean()) / df.iloc[:, 1:5].std()

# Print standardized scores
print(standardized_scores)
```

# 28
Visualize the total scores using a bar plot.

---

```python
plt.figure(figsize=(6, 4))
plt.bar(df["Student Name"], df["Total Score"], color=['green', 'black', 'purple', 'pink'])

# Add labels
plt.xlabel("Students")
plt.ylabel("Total Score")
plt.title("Total Scores of Students")
plt.show()
```

# 29
Generate a pie chart representing the subject-wise average scores.

---

```python
subject_means = df.iloc[:, 1:5].fillna(0).mean()

plt.figure(figsize=(6, 6))
plt.pie(subject_means, labels=subject_means.index, autopct='%1.1f%%', colors=['red', 'pink', 'yellow', 'green'])

# Add title
plt.title("Subject-wise Average Scores")
plt.show()
```

# 30
Display the students whose History score is greater than the average History score.

---

```python
# Compute average history score
avg_history = df["History"].mean()

# Filter students with above-average History scores
high_history_students = df[df["History"] > avg_history]
print(high_history_students[["Student Name", "History"]])
```
